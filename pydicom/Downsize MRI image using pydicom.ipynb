* <'pydicom.dataset.FileDataset' class>.pixel_array 
    * pydicom.dataset 모듈 내 Dataset 클래스를 상속 받은 FileDataset 클래스의 프로퍼티 
    * dcmread로 읽어들인 DICOM파일의 pixel 정보를 numpy.ndarray형태로 반환함 
* pixelData 수정
    * <'pydicom.dataset.FileDataset' class>.pixel_array로 얻은 <'numpy.ndarray' class> 수정 
    * <'pydicom.dataset.FileDataset' class>.pixelData에 대해 수정한 <'numpy.ndarray' class>.tobytes() 할당
    * <'pydicom.dataset.FileDataset' class>.Rows와 .Colums에 대해 수정한 <'numpy.ndarray' class>.shape 할당
* numpy.ndarray
    * shape : array dimensions 을 tuple 형태로 반환 (ex: 2차원이면  (Y픽셀수, X픽셀수) 로 반환)
* list 
    * list[n] : list의 n index의 값 (list아님. 1개값)
    * list[:n] : list의 0 idx ~ (n-1) idx 까지만의 sublist (list. length = n)
    * list[::n] : list에서 0 idx 부터 n칸씩 띄어가며 뽑은 수 sublist (list. length = 원래길이/n의 올림)
* 라이브러리 
    * numpy : pydicom pixel data 관련
    * Pillow : 
    * gdcm : miniconda (https://conda.io/projects/conda/en/latest/user-guide/install/macos.html) 를 설치하여 

```python
# authors : Guillaume Lemaitre <g.lemaitre58@gmail.com>
# license : MIT

import pydicom
from pydicom.data import get_testdata_files

print(__doc__)

# FIXME: add a full-sized MR image in the testing data
filename = get_testdata_files('MR_small.dcm')[0]
#filename = 'bmode.dcm'
ds = pydicom.dcmread(filename)
print(ds)

# get the pixel information into a numpy array
data = ds.pixel_array
print('The image has {} x {} voxels'.format(data.shape[0],
                                            data.shape[1]))
data_downsampling = data[::1, ::1]
print('The downsampled image has {} x {} voxels'.format(
    data_downsampling.shape[0], data_downsampling.shape[1]))

# copy the data back to the original data set
print(type(ds.PixelData))
ds.PixelData = data_downsampling.tobytes()
# update the information regarding the shape of the data array
ds.Rows, ds.Columns = data_downsampling.shape

# print the image information given in the dataset
print('The information of the data set after downsampling: \n')
print(ds)
```
